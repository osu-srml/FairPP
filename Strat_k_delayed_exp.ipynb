# Dynamics of Participation Rate of the minority group

- Use participation rate to measure fairness
- For DRO, fair with/without demographic attribute, get the participation rate dynamics for different $\rho$

```python
rrm_data = np.load(f'arrays/{name}_rrm_arrays.npy')
rrm_p0_mean = np.mean(rrm_data[1],axis=0)
rrm_p0_sd = np.std(rrm_data[1],axis=0)
dro_dict_mean, fair_dict_mean, fair_dict_mean2 = {0:rrm_p0_mean}, {0:rrm_p0_mean}, {0:rrm_p0_mean}
dro_dict_sd, fair_dict_sd, fair_dict_sd2 = {0:rrm_p0_sd}, {0:rrm_p0_sd}, {0:rrm_p0_sd}
rho_list = [0.3,0.7]
# read results if needed
for rho in rho_list:
    fair_data = np.load(f'arrays/{name}_fair_arrays_{rho}.npy')
    fair_p0_mean = np.mean(fair_data[1],axis = 0)
    fair_p0_sd = np.std(fair_data[1],axis = 0)
    fair_dict_mean[rho] = fair_p0_mean
    fair_dict_sd[rho] = fair_p0_sd

    fair_data2 = np.load(f'arrays/{name}_fair_arrays2_{rho}.npy')
    fair_p0_mean2 = np.mean(fair_data2[1],axis = 0)
    fair_p0_sd2 = np.std(fair_data2[1],axis = 0)
    fair_dict_mean2[rho] = fair_p0_mean2
    fair_dict_sd2[rho] = fair_p0_sd2

    dro_data = np.load(f'arrays/{name}_dro_arrays_{rho}.npy')
    p0_mean = np.mean(dro_data[1],axis=0)
    p0_sd = np.std(dro_data[1], axis=0)
    dro_dict_mean[rho] = p0_mean
    dro_dict_sd[rho] = p0_sd
```

```python
# Convergence Plot of Theta
# Create the figure and axis
rho_list = [0.3,0.7]
fig, ax = plt.subplots(nrows=1, ncols=3, figsize = (6,1.8))
plt.subplots_adjust(wspace=0.25) 
x = range(5,num_iter+1)
idx = range(4,num_iter)
x2 = range(3,num_iter+1)
idx2 = range(2,num_iter)
# colors = ['blue','#FFCC00','black','green','red']
colors = ['blue','black','red']
rho_unicode = '\u03C1'
rho_list = [0] + rho_list
for i in range(len(rho_list)):
    color = colors[i]
    rho = rho_list[i]
    ax[0].fill_between(x, dro_dict_mean[rho][idx] - dro_dict_sd[rho][idx], dro_dict_mean[rho][idx] + dro_dict_sd[rho][idx], color=color, alpha=0.2)
    ax[0].plot(x, dro_dict_mean[rho][idx], linestyle='-', linewidth = 1, color=color, label=f'{rho_unicode} = {rho}')
    ax[1].fill_between(x, fair_dict_mean[rho][idx] - fair_dict_sd[rho][idx], fair_dict_mean[rho][idx] + fair_dict_sd[rho][idx], color=color, alpha=0.2)
    ax[1].plot(x, fair_dict_mean[rho][idx], linestyle='-', linewidth = 1, color=color, label=f'{rho_unicode} = {rho}')
    ax[2].fill_between(x2, fair_dict_mean2[rho][idx2] - fair_dict_sd2[rho][idx2], fair_dict_mean2[rho][idx2] + fair_dict_sd2[rho][idx2], color=color, alpha=0.2)
    ax[2].plot(x2, fair_dict_mean2[rho][idx2], linestyle='-', linewidth = 1, color=color, label=f'{rho_unicode} = {rho}')

for a in ax:
    a.set_xlabel(r'$t$',fontsize=12)
    ax[0].set_ylabel(r'$P$',fontsize=12)

handles, labels = [], []
for handle, label in zip(*ax[0].get_legend_handles_labels()):
    handles.append(handle)
    labels.append(label)
    
fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 1.3), ncol=3, handlelength = 0.5)
plt.tight_layout()
fig.savefig(f'plots/{name}_prate.pdf',bbox_inches='tight')  
```
